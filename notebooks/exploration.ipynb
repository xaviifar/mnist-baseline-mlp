```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# Descargar dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
labels, counts = np.unique(y_train, return_counts=True)
```

```python
# Mostrar cuantos numeros hay de cada

fig, ax = plt.subplots()
ax.bar(x = labels, height = counts)
ax.set_xlabel("Etiqueta")
ax.set_ylabel("Número de muestras")
ax.set_title("Distribución de etiquetas en MNIST")

plt.show()
```

```python
# Mostrar imágenes del dataset

plt.figure(figsize=(6,6))

for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(y_train[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Mostrar una imagen de cada etiqueta del 0 al 9
plt.figure(figsize=(10, 5))
for i in range(10):
    # Obtener el índice de la primera ocurrencia del número i
    idx = np.where(y_train == i)[0][0]
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[idx], cmap='gray')
    plt.title(f"Dígito: {i}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
